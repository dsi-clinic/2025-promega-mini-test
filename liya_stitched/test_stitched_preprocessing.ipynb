# Test stitched image preprocessing

Two artifacts addressed:
1. The black corner patch
2. The red scale bar

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from stitched_image_preprocessing import remove_red_scalebar, remove_corner_blackbox

```

```python
# Real image data from promega (ba296_1_Dy30 has *_stitched.png files)
image_folder = '/net/projects2/promega/data-analysis/output/infer_resized_512x384/auto_processed/ba296_1_Dy30'

if not os.path.exists(image_folder):
    raise FileNotFoundError(f"Folder not found: {image_folder}")

png_files = sorted([f for f in os.listdir(image_folder) if f.lower().endswith('_stitched.png')])
print(f"Found {len(png_files)} stitched images\n")

for filename in png_files:
    img = Image.open(os.path.join(image_folder, filename))
    img.load()
    arr = np.array(img)

    img_remove_red = remove_red_scalebar(arr)
    img_remove_black = remove_corner_blackbox(img_remove_red)

    fig, ax = plt.subplots(2, 2, figsize=(12, 8))
    ax[0, 0].imshow(arr, interpolation='none')
    ax[0, 0].axis('off')
    ax[0, 0].set_title('Raw')
    fig.suptitle(filename, fontsize=14)
    ax[0, 1].imshow(img_remove_red, interpolation='none')
    ax[0, 1].set_title('Removing scale bar')
    ax[0, 1].axis('off')
    ax[1, 0].axis('off')
    ax[1, 1].imshow(img_remove_black, interpolation='none')
    ax[1, 1].set_title('Removing corner black boxes')
    ax[1, 1].axis('off')
    plt.show()
```
