```python
import json
import random
import tifffile
import matplotlib.pyplot as plt
from pathlib import Path

# --- Config ---
json_path = Path("/net/projects2/promega/data-analysis/output/image_mapping.json")
num_samples = 20

# --- Load JSON ---
with open(json_path) as f:
    mapping = json.load(f)

# --- Filter stitched entries ---
stitched_entries = {k: v for k, v in mapping.items() if v.get("Stitched") == "Yes"}

# --- Sample 20 stitched entries ---
sample_keys = random.sample(list(stitched_entries.keys()), min(num_samples, len(stitched_entries)))

for key in sample_keys:
    entry = stitched_entries[key]
    print(f"\n=== {key} ===")
    print(f"  Best Z File: {entry.get('Best Z Filename')}")
    print(f"  All Files: {entry.get('all_files')}")

    # Optional: visualize best Z-slice
    best_z_file = entry.get("Best Z Filename")
    if best_z_file and Path(best_z_file).is_file():
        img = tifffile.imread(best_z_file)
        plt.imshow(img, cmap='gray')
        plt.title(f"{key} - Best Z")
        plt.axis('off')
        plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

test_entry = stitched_entries[sample_keys[0]]
file_list = sorted(test_entry['all_files'])

arrays = []
heights = []

for f in file_list:
    img = Image.open(f).convert('L')  # convert to grayscale
    arr = np.array(img)
    print(f"{Path(f).name}: shape={arr.shape}, min={arr.min()}, max={arr.max()}")
    arrays.append(arr)
    heights.append(arr.shape[0])

# Pad arrays to same height
max_height = max(heights)
padded_arrays = []

for arr in arrays:
    h, w = arr.shape
    pad_top = (max_height - h) // 2
    pad_bottom = max_height - h - pad_top
    padded = np.pad(arr, ((pad_top, pad_bottom), (0, 0)), mode='constant', constant_values=0)
    padded_arrays.append(padded)

stitched = np.hstack(padded_arrays)

# Normalize for visualization
stitched_norm = (stitched - stitched.min()) / (stitched.max() - stitched.min() + 1e-5)
plt.imshow(stitched_norm, cmap='gray')
plt.title(f"{sample_keys[0]} - Stitched Horizontal")
plt.axis('off')
plt.show()

```
