```python
import json
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

# Path to all_data.json
json_path = Path("all_data.json")

# Load JSON
with open(json_path, "r") as f:
    data = json.load(f)

# Get all entries sorted by Blank flag
all_entries = [(k, v) for k, v in data.items()]
all_entries.sort(key=lambda x: not x[1].get("Blank", False))  # True first

# Limit number shown so plot isn’t huge
max_to_show = 20
entries_to_show = all_entries[:max_to_show]

# Prepare plot
# Prepare plot
cols = 5
rows = -(-len(entries_to_show) // cols)  # ceiling division
fig, axes = plt.subplots(rows, cols, figsize=(15, 3 * rows))

# Flatten axes array for easy iteration
axes = np.ravel(axes)

for ax, (img_id, info) in zip(axes, entries_to_show):
    img_path = Path(info["Best Z Filename"])
    if img_path.exists():
        img = cv2.imread(str(img_path))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        ax.imshow(img)
    else:
        ax.text(0.5, 0.5, "Missing", ha="center", va="center", fontsize=12)

    ax.set_title(f"{img_id}\nBlank={info.get('Blank', False)}", fontsize=8)
    ax.axis("off")

# Hide unused axes
for ax in axes[len(entries_to_show):]:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
